# Brian MF Parity Checks

This notebook compares selected `tvbtoolkit.brian_mf` outputs against the legacy TVBSim `brian_MF` implementation when available on disk.

```python
from pathlib import Path
import pandas as pd

from tvbtoolkit.brian_mf.parity.compare import compare_mf_with_legacy
from tvbtoolkit.brian_mf.parity.fixtures import fixed_fit_coefficients

LEGACY_ROOT = Path("/Users/borjan/CNRS/projects/TVBSim/brian_MF")
prs, pfs = fixed_fit_coefficients()
```

```python
if LEGACY_ROOT.exists():
    report = compare_mf_with_legacy(
        prs,
        pfs,
        legacy_root=LEGACY_ROOT,
        cells="FS-RS",
        amp_stim_hz=0.0,
        iext_hz=0.3,
        total_time_s=2.0,
        seed=123,
    )
    df = pd.DataFrame([report.__dict__])
    display(df)
else:
    print(f"Legacy checkout not found at: {LEGACY_ROOT}")
```
